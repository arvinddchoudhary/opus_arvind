2.	What is Linear Regression?

Linear Regression is one of the regression technique which is commonly used algorithm for regression. it tries to model the relationship between the independent variable and the dependent variable using a straight line.



3.	When to use Linear Regression? Explain the equation of a straight line.

Linear Regression is used when we want to predict a continuous numerical value and there is a linear relationship between input and the output. For example Linear regression can be used when a person salary is predicted using his previous years of experience

The goal of Linear Regression is to find the best fitting straight line that represents the relationship between the independent variable and the dependent variable.

    The equation of a simple linear regression line is:
    y=mx+b
    y = output/dependent variable/label
    x = input variable
    m = slope of the line
    b = intercept

5.	How is the best fit line chosen?

In Linear Regression, the best fit line is chosen by minimizing the difference between the actual values and the predicted values.

This difference is called the error or residual. The goal of the model is to find a line that makes these errors as small as possible.

Regression uses a method called Least Squares Methods. This method calculates the square of the error between actual data points and the predicted values from the line.
By minimizing the squared errors, the algorithm finds the values of the slope and intercept that produce the best possible line that fits the data.

7.	What are residuals?

residuals are the differences between the observed (actual) values in a dataset and the values predicted by a model (the fitted values)

    Mathematically, it can be written as:
    {Residual} = y - hat{y}
    y = actual value
    ŷ (y hat) = predicted value from the regression model

8.	What is correlation?

Correlation is a statistical measure that shows the strength and direction of the relationship between two variables.

The value of correlation ranges from -1 to +1.
+1 → Perfect positive correlation (both variables increase together)
-1 → Perfect negative correlation (one increases while the other decreases)

0 → No correlation

9.	What is multicollinearity?

Multicollinearity occurs when two or more independent variables are highly correlated with each other.

This makes it difficult for the model to know which variable is influencing the target.

    Problems caused:
    Unstable coefficients
    Poor model interpretation

10.	How to detect multicollinearity?

    Methods include:
    Correlation Matrix
    Heatmap(between -1.0 and 1.0)
    Variance Inflation Factor (VIF)

Rule of thumb:
VIF > 5 → moderate multicollinearity
VIF > 10 → severe multicollinearity

11.	What are the remedies for multicollinearity?

    Solutions include:
    Remove highly correlated features
    Use Principal Component Analysis (PCA)
    Use Regularization techniques (Ridge / Lasso /ElasticNet)
    Combine correlated variables

12.	What is the R-Squared Statistics?

R² measures how well the independent variables explain the variance in the dependent variable.

        R^2 = 1-(RSS/TSS)

    Range:
    0 → model explains nothing
    1 → perfect model

Example:
R² = 0.85 means 85% of variance is explained by the model.

13.	What is an adjusted R-Squared Statistics?

Adjusted R² is a modified version of R² that adjusts for the number of predictors in the model.
It penalizes unnecessary variables.

    Formula:
    Adjusted R^2=1−((1-R^2)(n-1)/(n-k-1))

    Where:
    n = number of observations
    k = number of predictors

14.	Why do we use adj R-squared?

Because R² always increases when new variables are added, even if they are useless.
Adjusted R² solves this by penalizing unnecessary variables.
Thus it gives a more reliable measure of model performance.

15.	Why adj R-squared decreases when we use incompetent variables?



If a new variable does not improve the model, the penalty term increases.
Therefore Adjusted R² decreases, indicating that the variable is not useful.

26.	Explain Ridge Regression.

Ridge Regression is a **regularization technique used to reduce overfitting in linear regression**. It adds a **penalty term equal to the square of the coefficients (L2 penalty)** to the cost function. This shrinks the coefficients but does not make them exactly zero.

Loss Function:
Loss = Sum of Squared Errors + λΣ(β²)

It helps in **handling multicollinearity and improving model generalization**.

27.	Explain Lasso Regression.

Lasso Regression (Least Absolute Shrinkage and Selection Operator) is a **regularization technique that adds an L1 penalty to the cost function**. It reduces the magnitude of coefficients and can **shrink some coefficients to exactly zero**, effectively performing **feature selection**.

Loss Function:
Loss = Sum of Squared Errors + λΣ|β|

It is useful when we want to **select the most important features**.

28.	Explain Elastic Net.

Elastic Net is a **combination of Ridge and Lasso regression**. It includes both **L1 and L2 regularization penalties** in the cost function.

Loss Function:
Loss = Sum of Squared Errors + λ₁Σ|β| + λ₂Σ(β²)
